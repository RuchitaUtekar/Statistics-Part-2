1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy import stats

sample_data = [110, 112, 115, 118, 120]
population_mean = 115
population_std_dev = 10

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}, P-value: {p_value:.4f}")

if p_value < 0.05:
   print("Reject the null hypothesis")
else:
   print("Fail to reject null hypothesis")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

sample_data = np.random.normal(loc=110, scale=15, size=50)
population_mean = 100
population_std_dev = 15

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}, Z-score: {z_score:.2f}, P-value: {p_value:.4f}")

x = np.linspace(population_mean - 4*population_std_dev, population_mean + 4*population_std_dev, 1000)
y = stats.norm.pdf(x, population_mean, population_std_dev / np.sqrt(len(sample_data)))

plt.plot(x, y, label='Normal Distribution', color = 'blue')
plt.fill_between(x, y, where=(x < population_mean - stats.norm.ppf(0.025)) | (x > population_mean + stats.norm.ppf(1-0.025)), color='red', alpha=0.5, label="Rejection Region")
plt.axvline(sample_mean, color = "green", linestyle="--", label=f"Sample Mean ({sample_mean:.2f})")
plt.axvline(population_mean - stats.norm.ppf(1-0.025), color="red", linestyle="--", label="Critical Value")
plt.axvline(population_mean + stats.norm.ppf(1-0.025), color="red", linestyle="--")


plt.title("Two-Tailed Z-Test")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

if p_value < 0.05:
   print("Reject the null hypothesis")
else:
   print("Fail to reject null hypothesis")
```

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

def visualize_error(population_mean=100, population_std=15, sample_mean=110, sample_std=15, n=30, alpha=0.05):
    std_error = population_std / np.sqrt(n)
    z_critical = norm.ppf(1- alpha / 2)
    critical_value_low = population_mean - z_critical * std_error
    critical_value_high = population_mean + z_critical * std_error
    x = np.linspace(population_mean - 4 * std_error, sample_mean + 4 * std_error, 1000)
    y_h0 = norm.pdf(x, population_mean, std_error)
    y_h1 = norm.pdf(x, sample_mean, std_error)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='Null Hypothesis', color='blue')
    plt.plot(x, y_h1, label='Alternate Hypothesis', color='red')
    plt.fill_between(x, y_h0, where=(x < critical_value_low) | (x > critical_value_high), color='red', alpha=0.4, label="Type 1 Error")
    plt.fill_between(x, y_h1, where=(x > critical_value_low) & (x < critical_value_high), color='orange', alpha=0.4, label="Type 2 Error")
    plt.axvline(critical_value_low, linestyle= '--' , color = 'black', label=f"Critical Values: {critical_value_low:.2f}, {critical_value_high:.2f}")
    plt.axvline(sample_mean, linestyle= '--' , color = 'green', label=f"Sample Mean: {sample_mean:.2f}")
    plt.title("Type 1 and Type 2 Errors")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()
visualize_error()
```

7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

def paired_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_rel(sample1, sample2)
    print("Paired T-test Results")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P_value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")

    plt.figure(figsize=(8,5))
    sns.boxplot(data=[sample1, sample2], palette=["blue", "red"])
    plt.xticks([0, 1], ["Before", "After"])
    plt.ylabel("Values")
    plt.title("Paired T-test")
    plt.grid(True)

    for i in range(len(sample1)):
       plt.plot([0,1], [sample1[i], sample2[i]], color="gray", alpha=0.6)

    plt.show()

np.random.seed(42)
before_scores = np.random.normal(loc=100, scale=15, size=30)
after_scores = before_scores + np.random.normal(loc=5, scale=5, size=30)

paired_t_test(before_scores, after_scores)

```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats
def calculate_confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1-alpha/2, df=sample_size - 1)
    standard_error = sample_std / np.sqrt(sample_size)
    margin_of_error = t_critical * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

lower,upper = calculate_confidence_interval(sample_data, confidence_level=0.95)
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f}) ")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayesian_inference(prior_h1, prior_h2, likelihood_h1, likelihood_h2):
    evidence = (likelihood_h1 * prior_h1) + (likelihood_h2 * prior_h2)
    posterior_h1 = ((likelihood_h1 * prior_h1)) / evidence
    posterior_h2 = ((likelihood_h1 * prior_h2)) / evidence
    return posterior_h1, posterior_h2

prior_h1 = 0.1
prior_h2 = 0.9
likelihood_h1 = 0.8
likelihood_h2 = 0.1

posterior_h1, posterior_h2 = bayesian_inference(prior_h1, prior_h2, likelihood_h1, likelihood_h2)
print(f"Posterior Probability for Hypothesis 1: {posterior_h1:.4f}")
print(f"Posterior Probability for Hypothesis 2: {posterior_h2:.4f}")

```

12.  Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

def chi_square_test(data):
    chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)
    print("Chi-square Test Results:")
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print("\nExpected Frequencies:")
    print(expected)

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")

data = np.array([[20, 15],
               [10, 25]])
df = pd.DataFrame(data, columns=["Category 1", "Category 2"], index=["Group 1", "Group 2"])
print("Contingency Table:")
print(df)

chi_square_test(df)
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
   observed_data = np.array(observed_data)
   row_sums = np.sum(observed_data, axis=1)
   col_sums = np.sum(observed_data, axis=0)
   total = np.sum(observed_data)
   expected_frequencies = total * np.outer(row_sums, col_sums) / total
   return expected_frequencies

observed_data = [[20, 15],
                [10, 25]]

expected = calculate_expected_frequencies(observed_data)
print("Observaed Data:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(np.round(expected, 2))

```

- The shape of the Chi-square distribution is determined by the degrees of freedom(df). As df increases, the distribution becomes more symmetric and resembles a normal distribution.
- For smaller degrees of freedom the distribution is heavily skewed to the right.
- The Chi-square distribution only takes non-negative values, ranging from 0 to infinity.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_Chi_square_distribution(degrees_of_freedom, sample_size=10000):
    chi_square_samples = np.random.chisquare(df=degrees_of_freedom, size=sample_size)

    plt.figure(figsize=(10, 6))
    sns.histplot(chi_square_samples, bins=30, kde=True, color="blue", stat="density")
    plt.title(f"Chi-square Distribution (df={degrees_of_freedom})")
    plt.xlabel("Chi-square Value")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

visualize_Chi_square_distribution(degrees_of_freedom=2)
visualize_Chi_square_distribution(degrees_of_freedom=5)
visualize_Chi_square_distribution(degrees_of_freedom=10)
```

16.  Implement an F-test using Python to compare the variances of two random samples.


```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
    else:
        f_stat = var2 / var1

    dfn = len(sample1) - 1
    dfd = len(sample2) - 1

    p_value = 1 - stats.f.cdf(f_stat, dfn, dfd)

    return f_stat, p_value

np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=10, scale=3, size=50)

f_stat, p_value = f_test(sample1, sample2)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
   print("Reject the null hypothesis")
else:
   print("Fail to reject the null hypothesis")

```

19.  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions (groups):
    print("Normality Test(Shapiro-Wilk):")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Shapiro-Wilk stat = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i}: Data is not normally distributed")
        else:
            print(f"Group {i}: Data is normally distributed")
    print("\nEqual Variance Test (Levene's Test):")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's stat = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("The variances are not equal")
    else:
        print("The variances are equal")
    print("\nIndependence Assumption:")
    print("Independence assumption is satisfied")

def plot_group_distribution(groups):
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=groups, palette="Set3")
    plt.title("Group Distributions")
    plt.xlabel("Group")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 6))
    for i, group in enumerate(groups, 1):
        plt.subplot(1, len(groups), i)
        plt.hist(group, bins=10, color='blue', edgecolor="black", alpha=0.7)
        plt.title(f"Group {i} Distribution")
        plt.xlabel("Value")
        plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

group1 = np.array([25, 30, 28, 21, 18])
group2 = np.array([35, 36, 33, 32, 31])
group3 = np.array([19, 22, 20, 17, 16])

groups = [group1, group2, group3]

check_anova_assumptions(groups)
plot_group_distribution(groups)
```

23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def hypothesis_test(sample_data, population_mean, alpha=0.05):
    t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis")

np.random.seed(42)
sample_size = 50
sample_mean = 100
sample_std = 15
sample_data = np.random.normal(loc=sample_mean, scale=sample_std, size=sample_size)

population_mean = 105
hypothesis_test(sample_data, population_mean)

plt.figure(figsize=(8, 6))
plt.hist(sample_data, bins=10, color='blue', edgecolor="black", alpha=0.7)
plt.axvline(np.mean(sample_data), color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean ({np.mean(sample_data):.2f})')
plt.axvline(population_mean, color='green', linestyle='dashed', linewidth=2, label=f'Population Mean ({population_mean})')
plt.title("Sample Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_test(sample_data, population_variance, alpha=0.05):
    n = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)
    chi_square_stat = (n - 1) * sample_variance / population_variance
    df = n - 1
    chi_critical_low = stats.chi2.ppf(alpha / 2, df)
    chi_critical_high = stats.chi2.ppf(1 - alpha / 2, df)
    p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))
    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Chi-square Statistic: {chi_square_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"Critical Values: {chi_critical_low:.4f}, {chi_critical_high:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
      print("\nFail to reject the null hypothesis")

    return chi_square_stat, df

np.random.seed(42)
sample_size = 30
true_variance = 20
sample_std = np.sqrt(true_variance)
sample_data = np.random.normal(loc=50, scale=sample_std, size=sample_size)

hypothesized_variance = 25
chi_square_stat, df = chi_square_test(sample_data, hypothesized_variance)

x = np.linspace(stats.chi2.ppf(0.001, df), stats.chi2.ppf(0.999, df), 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(8,6))
plt.plot(x, y, label=f"Chi-Square Distribution (df={df})", color='blue')
plt.axvline(chi_square_stat, color ='red', linestyle='dashed', linewidth=2, label=f"Test Statistic: {chi_square_stat:.2f}")
plt.axvline(stats.chi2.ppf(0.05 / 2, df), color='green', linestyle='dashed', label=f"Critical Value(Lower)")
plt.axvline(stats.chi2.ppf(1 - 0.05 / 2, df), color='green', linestyle='dashed', label=f"Critical Value(Upper)")
plt.title("Chi-Square Test")
plt.xlabel("Chi-square Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

25.  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats

def z_test_proportion(successes1, n1, successes2, n2, alpha=0.05):
    p1 = successes1 / n1
    p2 = successes2 / n2
    p_pooled = (successes1 + successes2) / (n1 + n2)
    standard_error = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n1 + 1 / n2))
    z_score = (p1 - p2) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    print(f"Proportion in Group 1: {p1:.4f}")
    print(f"Proportion in Group 2: {p2:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")

successes_group1 = 45
sample_size_group1 = 200
successes_group2 = 55
sample_size_group2 = 220

z_test_proportion(successes_group1, sample_size_group1, successes_group2, sample_size_group2)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * min(stats.f.cdf(f_stat, dfn, dfd), 1 - stats.f.cdf(f_stat, dfn, dfd))

    print(f"Variance of Sample1 : {var1:.4f}")
    print(f"Variance of Sample2 : {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P_value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null")

np.random.seed(42)
sample_size = 30
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

f_test(sample1, sample2)
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
sns.boxplot(data=[sample1, sample2], palette="Set2")
plt.xticks([0, 1], ["Sample 1", "Sample 2"])
plt.title("Boxplot of the two samples")

plt.subplot(1, 2, 2)
plt.hist(sample1, bins=10, alpha=0.6, color='blue', edgecolor="black", label= "Sample 1")
plt.hist(sample2, bins=10, alpha=0.6, color='red', edgecolor="black", label= "Sample 2")
plt.title("Histogram of the two samples")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```

27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit(observed, expected, alpha=0.05):

    chi2_stat, p_value = stats.chisquare(observed, expected)

    print(f"Chi-Square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("\nReject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
    else:
        print("\nFail to reject the null hypothesis: The observed distribution matches the expected distribution.")

np.random.seed(42)
observed_counts = np.random.multinomial(n=120, pvals=[1/6]*6)

expected_counts = [120 / 6] * 6

chi_square_goodness_of_fit(observed_counts, expected_counts)

labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, observed_counts, width=0.4, label="Observed", color='blue', alpha=0.7)
plt.bar(x + 0.2, expected_counts, width=0.4, label="Expected", color='orange', alpha=0.7)
plt.xticks(x, labels)
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.title("Observed vs Expected Frequencies (Dice Rolls)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
